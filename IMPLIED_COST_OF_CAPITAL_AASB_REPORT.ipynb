```python
#@title RETRIEVE MARKET DATA AND IMPLIED COST OF EQUITY FOR A GROUP OF FIRMS USING TICKER INPUT. EXAMPLE: ANZ,CBA,BEN,BOQ,NAB,MQG,SUN,WBC

import yfinance as yf
import datetime
tick=input("What firms would you like to derive the implied cost of capital, seperated by comma (no space)? \n")
names=tick.split(",")
# names=["ANZ","CBA","BEN","BOQ","NAB","VUK",
#        "SUN","WBC"]
names = list(dict.fromkeys(names))
print(names)
today = datetime.date.today()

beta=[]
marketcap=[]
cost=[]
year=[]
compname=[]
sector=[]
for name in names:
    firm =yf.Ticker(f"{name}.AX")
    # print(f"{name} ROE is {firm.info['returnOnEquity']}")
    # print(f"{name} Book Value is {firm.info['bookValue']}")
    # print(f"{name} Market Price is $ {firm.info['regularMarketPrice']}")


    B=firm.info['bookValue']
    g=0.025
    ROE=firm.info['returnOnEquity']
    V=firm.info['currentPrice']
    #Seed discount value
    r=g+0.000001
    TEMP=round(B+((ROE-r)/(r-g))*B,2)
    diff=round(abs(V-TEMP),2)
    # print(TEMP,diff)
    beta.append(firm.info['beta'])
    marketcap.append(round(firm.info['marketCap']/1000000000,2))
    
    year.append(today.year)
    compname.append(firm.info['longName'])
    sector.append(firm.info['sector'])

    iter=0

    while diff>0.0001 and TEMP>V:
        r+=0.00001
        TEMP=round(B+((ROE-r)/(r-g))*B,2)
        diff=round(abs(V-TEMP),2)
        #print(r,TEMP,diff)
        # cost="{:.3%}".format(r)
        iter+=1
    
       
    cost.append(r) 

    print(f"The implied cost of equity capital of {name} is {r: .2%} after {iter} iterations to derive a market value of $ {TEMP} per share")

```
