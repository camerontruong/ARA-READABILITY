```python
#RETRIEVE FINANCIAL DATA FROM YAHOO FINANCE FOR ASX LISTED FIRMS

import yfinance as yf
tick=input("What firm would you like to derive the implied cost of capital? \n").upper()
firm =yf.Ticker(f"{tick}.AX")
B=firm.info['bookValue']
g=0.025
ROE=firm.info['returnOnEquity']
V=firm.info['currentPrice']
#Seed discount value
r=g+0.000001
TEMP=round(B+((ROE-r)/(r-g))*B,2)
diff=round(abs(V-TEMP),2)
# print(TEMP,diff)
mcap=round(firm.info['marketCap']/1000000000,2)
print(f"{tick} company name is {firm.info['longName']}")
print(f"{tick} is the sector of {firm.info['sector']}")
print(f"{tick} Market Cap (in billion) is $ {mcap}")
print(f"{tick} Market Beta is {round(firm.info['beta'],2)}")
print(f"{tick} ROE is {round(firm.info['returnOnEquity'],4): .2%}")
print(f"{tick} Book Value is {firm.info['bookValue']}")
print(f"{tick} Market Price is $ {firm.info['currentPrice']}")
# print(f"{tick} Market Cap is $ {firm.info['marketCap']}")
# print(mcap)
# print(firm.info)
```

```python
#EMPLOY ITERATIONS TO DERIVE THE DISCOUNT RATE FOR THE ONE-STAGE RESIDUAL INCOME MODEL
iter=0

while diff>0.0001 and TEMP>V:
    r+=0.00001
    TEMP=round(B+((ROE-r)/(r-g))*B,2)
    diff=round(abs(V-TEMP),2)
    #print(r,TEMP,diff)
    cost="{:.3%}".format(r)
    iter+=1
    print(f"This is the {iter} iteration, discount rate used is {r: .3%}, valuation is $ {TEMP}, market price is $ {V} and difference is $ {diff}")

print(f"The implied cost of capital of {tick} is {r: .2%} after {iter} iterations to derive a market value of $ {TEMP}")
```

```python
# ALLOW a PANEL of FIRMs
#FULL RUN
tick=input("What firms would you like to derive the implied cost of capital, seperated by comma? \n")
names=tick.split(",")
# names=["ANZ","NAB","CBA","BOQ","BHP","FMG","RIO","JHX","BSL","NCM","NST","CSL","BPT","STO","WPL","RMD","NCK","SUL","TWE","WOW","COL",
#        "SUN","ALL","GNC","CGF","RHC","AWC","RRL","NHC","ARB","APX","RFF","TLS","IGO","WTC","LLC","WES","LYC","EVN","IEL","JBH","HVN",
#        "NUF","MIN","DOW","A2M","SSM","AIA","AMC","ASX","COH","FPH","MGR","REA","REH","SEK","SHL",
#        "SGP","TAH","WBC"]
names = list(dict.fromkeys(names))
print(names)


beta=[]
marketcap=[]
cost=[]
year=[]
compname=[]
sector=[]
for name in names:
    firm =yf.Ticker(f"{name}.AX")
    # print(f"{name} ROE is {firm.info['returnOnEquity']}")
    # print(f"{name} Book Value is {firm.info['bookValue']}")
    # print(f"{name} Market Price is $ {firm.info['regularMarketPrice']}")


    B=firm.info['bookValue']
    g=0.025
    ROE=firm.info['returnOnEquity']
    V=firm.info['regularMarketPrice']
    #Seed discount value
    r=g+0.000001
    TEMP=round(B+((ROE-r)/(r-g))*B,2)
    diff=round(abs(V-TEMP),2)
    # print(TEMP,diff)
    beta.append(firm.info['beta'])
    marketcap.append(round(firm.info['marketCap']/1000000000,2))
    year.append(2022)
    compname.append(firm.info['longName'])
    sector.append(firm.info['sector'])

    iter=0

    while diff>0.0001 and TEMP>V:
        r+=0.00001
        TEMP=round(B+((ROE-r)/(r-g))*B,2)
        diff=round(abs(V-TEMP),2)
        #print(r,TEMP,diff)
        # cost="{:.3%}".format(r)
        iter+=1
    
       
    cost.append(r) 

    print(f"The implied cost of capital of {name} is {r: .2%} after {iter} iterations to derive a market value of $ {TEMP}")

```
